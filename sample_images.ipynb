```python
# Dependencies
import torch
from torch import nn
from torchvision import transforms, datasets
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

# Reading data
# images dir
train_dir = 'data/classes/'

# define transforms: resize the image to 3840 x 2160
transform = transforms.Compose([transforms.Resize((2160, 3840)), transforms.ToTensor()])
# load data
train_data = datasets.ImageFolder(train_dir, transform=transform)

# define dataloaders
trainloader = torch.utils.data.DataLoader(train_data, batch_size=64, shuffle=True)

image, label = trainloader.dataset[110]
# show the image
plt.imshow(image.permute(1, 2, 0))
plt.axis('off')
plt.show()
print("Class: ", train_data.classes[label])
```
